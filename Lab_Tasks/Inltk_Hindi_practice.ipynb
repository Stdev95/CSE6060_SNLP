```python
from inltk.inltk import identify_language
```

```python
identify_language(file)
```

```python
from inltk.inltk import predict_next_words
```

```python
sentence = "बहुत समय से मिले नहीं"
```

```python
predict_next_words(sentence,5,'hi')
```

```python
from inltk.inltk import get_similar_sentences
```

```python
get_similar_sentences(sentence,5,'hi')
```

```python
get_similar_sentences('मैं आज बहुत खुश हूं', 5, 'hi')
```
